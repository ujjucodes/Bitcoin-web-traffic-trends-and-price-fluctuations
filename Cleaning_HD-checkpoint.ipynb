#### Dataset Source and Description

The dataset used in this project was sourced from [Stack Exchange Data Dump](https://archive.org/details/stackexchange), a publicly available repository that provides anonymized data from the Stack Exchange network. This dataset encompasses a wide range of community-driven Q&A sites and includes detailed records on user activities such as posts, comments, badges, votes, and user profiles.

The data is originally stored in XML files, each representing a specific aspect of the platform—for example, `Posts.xml` contains all questions and answers, while `Users.xml` includes user profiles. These XML files are rich in metadata and structured attributes that capture the dynamics of user engagement on the platform over time.

#### Code Functionality

The provided Python code automates the process of converting these XML files into CSV format, which is more accessible and easier to manipulate for data analysis tasks. Using Python’s `xml.etree.ElementTree` library, the code parses each XML file, extracts the attributes of each record, and converts them into tabular form with pandas DataFrames. These DataFrames are then saved as CSV files for efficient storage and subsequent analysis.

Furthermore, the code loads each generated CSV into a pandas DataFrame, storing them in a dictionary for organized access during analysis. This pipeline facilitates seamless data cleaning, transformation, and visualization, enabling the exploration of user behavior and content trends across the Stack Exchange communities.


```python
# Function to convert XML to CSV
def xml_to_csv(xml_file, csv_file):
    try:
        tree = ET.parse(xml_file)
        root = tree.getroot()

        data = [child.attrib for child in root]
        df = pd.DataFrame(data)
        df.to_csv(csv_file, index=False)
        print(f"Converted: {xml_file} → {csv_file}")
    except Exception as e:
        print(f"Error processing {xml_file}: {e}")

# List of all XML base names (without .xml extension)
file_names = ["Badges", "Comments", "PostHistory", "PostLinks", "Posts", "Tags", "Users", "Votes"]

# Convert each XML to CSV
for name in file_names:
    xml_file = f"{name}.xml"
    csv_file = f"{name}.csv"
    xml_to_csv(xml_file, csv_file)

# Load CSVs into pandas DataFrames (stored in a dictionary for easy access)
dataframes = {}
for name in file_names:
    csv_file = f"{name}.csv"
    try:
        df = pd.read_csv(csv_file)
        dataframes[name] = df
        print(f"Loaded: {csv_file}")
    except Exception as e:
        print(f"Error loading {csv_file}: {e}")
```

### Null Value Handling

This code performs data cleaning and handling of missing values in the `posts` DataFrame:

- Rows with missing values in critical columns `'Body'` and `'PostTypeId'` are removed, as these are essential for analysis.
- Missing values in numerical columns like `'ViewCount'`, `'FavoriteCount'`, and `'AnswerCount'` are filled with `0` because absence of data in these fields likely means no views, favorites, or answers.
- Missing identifiers in `'AcceptedAnswerId'` and `'ParentId'` are replaced with `-1` to indicate no accepted answer or parent post, keeping the column numeric while clearly marking missing data.
- For textual columns like `'Tags'`, `'Title'`, and `'Body'`, missing values are filled with placeholder strings (`'no-tag'`, `'no-title'`, `'no-body'`) to avoid issues during text processing and to explicitly mark missing content.

These steps ensure the dataset is complete and consistent, enabling accurate analysis and visualization without errors caused by missing values.
#### For Posts

```python
posts = posts.dropna(subset=['Body', 'PostTypeId'])

posts['ViewCount'] = posts['ViewCount'].fillna(0)
posts['FavoriteCount'] = posts['FavoriteCount'].fillna(0)
posts['AnswerCount'] = posts['AnswerCount'].fillna(0)
posts['AcceptedAnswerId'] = posts['AcceptedAnswerId'].fillna(-1)
posts['ParentId'] = posts['ParentId'].fillna(-1)

posts['Tags'] = posts['Tags'].fillna('no-tag')
posts['Title'] = posts['Title'].fillna('no-title')
posts['Body'] = posts['Body'].fillna('no-body')
```

In this data cleaning process, missing values across multiple datasets are handled to ensure consistency and avoid errors during analysis. For the **tags** dataset, missing IDs are replaced with `-1` to signify unknown or absent references. 

In the **comments** dataset, empty comment texts are filled with `'no-comment'`, and missing user information is replaced with `'Anonymous'` or `-1`, indicating anonymous or unknown users.

The **post history** dataset’s missing text and comments are similarly filled with placeholder strings to preserve the structure and meaning of the data. 

For **post links**, missing related post IDs and link types are replaced with `-1`, indicating undefined relationships.

Finally, in the **votes** dataset, user IDs and bounty amounts are filled with default values if those columns exist, ensuring numerical operations can be performed without interruption. Overall, these steps help maintain data integrity and allow for smooth downstream processing and analysis.


```python
# Fill missing values in tags dataset to avoid NaNs
tags['ExcerptPostId'] = tags['ExcerptPostId'].fillna(-1)
tags['WikiPostId'] = tags['WikiPostId'].fillna(-1)

# Fill missing comment texts and user info to maintain data integrity
comments['Text'] = comments['Text'].fillna('no-comment')
comments['UserDisplayName'] = comments['UserDisplayName'].fillna('Anonymous')
comments['UserId'] = comments['UserId'].fillna(-1)

# Handle missing text and comment fields in post history
posthistory['Text'] = posthistory['Text'].fillna('no-text')
if 'Comment' in posthistory.columns:
    posthistory['Comment'] = posthistory['Comment'].fillna('no-comment')
posthistory['UserId'] = posthistory['UserId'].fillna(-1)

# Fill missing values in post links dataset
postlinks['RelatedPostId'] = postlinks['RelatedPostId'].fillna(-1)
postlinks['LinkTypeId'] = postlinks['LinkTypeId'].fillna(-1)

# Handle missing user and bounty info in votes dataset if columns exist
if 'UserId' in votes.columns:
    votes['UserId'] = votes['UserId'].fillna(-1)
if 'BountyAmount' in votes.columns:
    votes['BountyAmount'] = votes['BountyAmount'].fillna(0)
```

#### Visual 1-Tag Extraction and Word Cloud Generation

This block of code focuses on visualizing the most common tags used in posts by generating a word cloud. The process starts by accessing the `Tags` column from the `posts_df` DataFrame, which contains tag information stored in a pipe-separated format (e.g., `|python|pandas|`). First, it removes any missing (`NaN`) entries and strips the leading and trailing pipes from each tag string. Then, it splits each tag string into a list of individual tags.

These lists of tags are flattened into a single list using list comprehension, resulting in one long list of all tags from all posts. This list is then joined into a single string where tags are separated by spaces — a format required by the `WordCloud` library to generate a word cloud.

A word cloud is then generated using `WordCloud()` with specified dimensions and color scheme. The result is a visual representation where frequently used tags appear larger, allowing quick identification of popular topics and areas of interest in the dataset. This visualization provides intuitive insights into what technologies or themes are most discussed within the community.


```python
# Remove NaN, remove leading/trailing |, then split by |
all_tags = posts_df['Tags'].dropna().apply(lambda x: x.strip('|').split('|'))

# Flatten list of lists into one list
flat_tags = [tag for sublist in all_tags for tag in sublist]

# Join into one string with spaces (wordcloud expects a string)
tag_text = ' '.join(flat_tags)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(tag_text)

# Plot the word cloud
plt.figure(figsize=(15,7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### Visual 2- Deep Dive into Tag Frequency: Top 20 Most Frequent Tags

This section extends the word cloud analysis by quantitatively exploring the most commonly used tags in the dataset. While the word cloud provides a visual representation of tag prominence, this code block dives deeper by calculating and visualizing the exact count of tag occurrences.

The process begins by ensuring that all entries in the `Tags` column of `posts_df` are treated as strings and removing any missing values. Each tag string is cleaned by removing the surrounding pipe (`|`) characters and then split into individual tags using a helper function.

All tags from all posts are collected into a single list, which is then passed through Python’s `Counter` from the `collections` module to calculate the frequency of each unique tag. The top 20 most frequent tags are extracted and prepared for visualization.

Finally, a horizontal bar chart is plotted using `seaborn` to clearly show the count of occurrences for each of these popular tags. This quantitative analysis complements the word cloud by providing precise numbers and revealing which technologies or topics dominate user discussions.


```python
# Ensure all tags are strings and drop missing values
tags_series = posts_df['Tags'].dropna().astype(str)

# Function to clean and split tags
def clean_tags(tag_str):
    return tag_str.strip('|').split('|')

# Flatten tag list
all_tags = []
for tag_str in tags_series:
    all_tags.extend(clean_tags(tag_str))

# Count frequencies
tag_counts = Counter(all_tags)
top_tags = tag_counts.most_common(20)
tags, counts = zip(*top_tags)

# Plot bar chart
plt.figure(figsize=(14, 7))
sns.barplot(x=list(counts), y=list(tags), palette='viridis')
plt.xlabel("Number of Occurrences")
plt.ylabel("Tags")
plt.title("Top 20 Most Frequent Tags")
plt.tight_layout()
plt.show()
```

### Visual 3- Temporal Analysis: Number of Posts Created Per Year

This section performs a time-based analysis to understand the yearly activity trends on the platform by examining how many posts were created each year. The process begins by ensuring that the `CreationDate` column in the `posts_df` DataFrame is converted into a proper datetime format using `pd.to_datetime()`. Any values that cannot be parsed into valid dates are safely coerced to `NaT` (Not a Time).

Once properly formatted, the `Year` is extracted from each post’s creation date and stored in a new column. This allows the dataset to be grouped by year using the `groupby()` function, where `.size()` counts the total number of posts per year.

The grouped data is then visualized using a bar chart to give a clear representation of platform activity over time. Each bar indicates the number of posts made in a particular year, offering insights into the growth and engagement of the community. For example, a sharp rise in post count might indicate increased interest in Bitcoin or related technologies during a particular period.

The visualization is styled for readability, with custom figure size, rotated x-axis labels, and grid lines to help highlight year-to-year differences. This analysis not only reveals usage patterns but also reflects user engagement and potential spikes in public interest or significant events within the dataset's timeline.


```python
# Ensure CreationDate is datetime
posts_df['CreationDate'] = pd.to_datetime(posts_df['CreationDate'], errors='coerce')

# Extract year
posts_df['Year'] = posts_df['CreationDate'].dt.year

# Count posts per year
posts_per_year = posts_df.groupby('Year').size()

# Plot
plt.figure(figsize=(12,6))
posts_per_year.plot(kind='bar', color='skyblue')
plt.title('Number of Posts Created Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Posts')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

To deepen my analysis of post activity trends on the platform, particularly during years when the number of posts peaked, I extended my exploration by incorporating external economic indicators. Cryptocurrency trends often trigger spikes in user interest, leading to a higher number of posts, questions, and discussions on tech and finance forums.

To investigate this, I loaded historical Bitcoin price data from a CSV file (`BitcoinData.csv`) and prepared it for visualization. The first step involved parsing the 'Date' column correctly using `pd.to_datetime`, ensuring compatibility regardless of the date format. This was crucial for aligning the Bitcoin price timeline with post activity records. 

Next, I cleaned the numeric columns (`Open`, `High`, `Low`, and `Price`) by removing commas and converting the values to float types. This data cleaning ensured that all values could be accurately processed and visualized without errors.

The dataset was then sorted in chronological order to maintain temporal accuracy. I used Plotly’s `go.Candlestick` chart to visualize the fluctuations in Bitcoin prices over time. The candlestick chart effectively highlights daily price movements by showing the open, high, low, and close prices. Green candles indicate price increases, while red candles signify decreases.




```python
# Load data
bitcoin_df = pd.read_csv('BitcoinData.csv')

# Parse date correctly (weekly data, dayfirst=True if dates are like DD/MM/YYYY)
bitcoin_df['Date'] = pd.to_datetime(bitcoin_df['Date'], dayfirst=True, errors='coerce')

# Clean numeric columns: remove commas and convert to float
for col in ['Open', 'High', 'Low', 'Price']:
    bitcoin_df[col] = bitcoin_df[col].astype(str).str.replace(',', '').astype(float)

# Sort by Date ascending
bitcoin_df = bitcoin_df.sort_values('Date')

# Plot candlestick
fig = go.Figure(data=[go.Candlestick(
    x=bitcoin_df['Date'],
    open=bitcoin_df['Open'],
    high=bitcoin_df['High'],
    low=bitcoin_df['Low'],
    close=bitcoin_df['Price'],
    increasing_line_color='green',
    decreasing_line_color='red',
    name='Bitcoin Price'
)])

fig.update_layout(
    title='Bitcoin Price Candlestick Chart',
    xaxis_title='Date',
    yaxis_title='Price (USD)',
    xaxis_rangeslider_visible=True,
    template='plotly_dark',
    height=600,
    width= 900
)


fig.show()
```

```python
# Convert CreationDate to datetime
posts_df['CreationDate'] = pd.to_datetime(posts_df['CreationDate'], errors='coerce')

# Drop rows with missing CreationDate
posts_df = posts_df.dropna(subset=['CreationDate'])

# Extract Hour and Weekday
posts_df['Hour'] = posts_df['CreationDate'].dt.hour
posts_df['Weekday'] = posts_df['CreationDate'].dt.day_name()

# Group and count posts by weekday and hour
heatmap_data = posts_df.groupby(['Weekday', 'Hour']).size().unstack(fill_value=0)

# Optional: Order the days
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
heatmap_data = heatmap_data.reindex(days_order)

# Plot heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=0.5)
plt.title('Post Activity Heatmap by Day and Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.show()
```

The heatmap above reveals when people are most actively posting or searching about Bitcoin-related topics throughout the week. One clear observation is that activity peaks during the later hours of the day, particularly between **2 PM and 9 PM**, with the highest activity reaching over **640 posts/hour** on **Wednesdays at 4 PM**.

### Visual 7- User Distribution on World Map (Log Scale)

This code analyzes the geographical distribution of users based on their self-reported locations in the `users_df` DataFrame. The first step involves defining a function, `extract_country`, which scans the `Location` string for the presence of any recognized country names using the `pycountry` library. If a country name is found, it returns the standardized country name; otherwise, it returns `None`. This approach helps to standardize diverse location inputs into consistent country labels, though it depends on the location strings containing recognizable country names.

The function is then applied to the `Location` column to create a new `Country` column. The code proceeds by counting the number of users from each country and storing these counts in a DataFrame, `country_counts`, with columns `Country` and `UserCount`.

To better visualize the data, especially since user counts often vary widely across countries, a logarithmic transformation (`log1p`) is applied to the counts to reduce skewness and highlight contrasts between countries with small and large user bases.

The core visualization is a choropleth world map created with Plotly Express (`px.choropleth`). It colors countries based on the log-transformed user counts, using the 'Plasma' color scale for aesthetic clarity. The color bar is customized to display readable tick marks corresponding to actual user counts (e.g., 1, 10, 100, 1k, 10k), improving interpretability.

Finally, the map layout is refined to enhance presentation by adjusting margins for better centering, enabling coastlines, and setting an equirectangular projection for a balanced world view. This visualization provides a clear, insightful overview of where the platform’s users are concentrated globally, which is valuable for understanding geographic trends, regional engagement, and potential areas for growth or outreach.


```python
# Extract country from location
def extract_country(location):
    for country in pycountry.countries:
        if country.name.lower() in location.lower():
            return country.name
    return None

# Apply extraction
users_df['Country'] = users_df['Location'].dropna().apply(extract_country)

# Count users by country
country_counts = users_df['Country'].value_counts().reset_index()
country_counts.columns = ['Country', 'UserCount']

# Add a log-transformed column for better contrast
country_counts['LogUserCount'] = np.log1p(country_counts['UserCount'])

# Plot choropleth using log-scaled values
fig = px.choropleth(
    country_counts,
    locations='Country',
    locationmode='country names',
    color='LogUserCount',
    hover_name='Country',
    title='User Distribution by Country (Log Scale)',
    color_continuous_scale='Plasma',
    height=700,
    width=1100
)

# Show proper legend with readable values
fig.update_traces(
    colorbar=dict(
        title='Log(User Count)',
        titleside='right',
        ticks='outside',
        tickvals=[np.log1p(x) for x in [1, 10, 100, 1000, 10000]],
        ticktext=['1', '10', '100', '1k', '10k']
    )
)

# Center the map using tighter and even margins
fig.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type='equirectangular'
    ),
    margin=dict(l=50, r=50, t=60, b=50),  # Balanced margins to center it
    title_x=0.5  # Center the title as well
)

fig.show()
```

### Data observation- Geographic Distribution of Users

The choropleth map above visualizes **user activity by country** on a log scale, indicating where Bitcoin-related searches or threads are most frequently generated. As expected, **first-world countries** like the **United States, Canada, Germany, the UK, and Australia** show significantly higher engagement, shaded in bright yellow or orange. These nations tend to have better internet infrastructure, higher financial literacy, and more disposable income to explore or invest in crypto technologies.

In contrast, many **African countries and less-developed regions** appear in deep purple or black, suggesting minimal search activity. This could be due to limited internet access, lack of awareness, or economic priorities that differ from crypto engagement.

Among the most active countries are:
- **United States** (highest)
- **India**, **Germany**, **Australia**, and **Canada** (high engagement)
- **Nigeria**, **Ethiopia**, and parts of **Central Africa** (lowest engagement)

**Australia** stands out with **moderate-to-high activity**, shaded orange, placing it **above median** in terms of global engagement.
